```python
import pandas as pd

# Load the CSV
df = pd.read_csv("MBTI_Dataset.csv")

# Take first and second columns by position
col0 = df.iloc[:, 0].astype(str)
col1 = df.iloc[:, 1].astype(str)

# Combine them
combined = col0 + ": " + col1

# Turn into one big training string (optional)
text = "\n".join(combined)

output_path = "MBTI_translated.txt"

with open(output_path, "w", encoding="utf-8") as f:
    f.write("\n".join(combined))

print(f"Wrote {len(combined)} lines to {output_path}")
```

```python
# Open the csv file
with open ('MBTI_Dataset.csv', 'r', encoding = 'utf-8') as f:
    text= f.read()

print ("length of dataset in characters: ", len(text))

```

```python
# Let's look at the first 1000 characters
print(text[:1000])
```

```python
chars = sorted(list(set(text))) # Creating a list of all of the unique characters in the dataset, which will then be sorted
vocab_size = len(chars) # Number of them
print(' '.join(chars))
print(vocab_size)
```

```python
stoi = {ch:i for i, ch, in enumerate(chars)}
itos = {i:ch for i, ch, in enumerate (chars)}
encode = lambda s: [stoi[c] for c in s] #encoder: take a string, output a list of integers
decode = lambda l: ''.join([itos[i] for i in l]) #decoder, take a list of integers, output a string

print (encode("greetings"))
print(decode(encode("greetings")))
```
